```python
import numpy as np
import matplotlib.pyplot as plt

# Configuration for bandwidth and traffic
bw_low = 20
bw_high = 2500
traffic_low = 1
traffic_high = 51
number_of_clients = 50

# Generate random data for bandwidth and traffic
bandwidths = np.random.randint(bw_low, bw_high + 1, size=number_of_clients)
traffic = np.random.randint(traffic_low, traffic_high, size=number_of_clients)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the data points
plt.scatter(traffic, bandwidths, s=100, edgecolor='orange', facecolor='none', linewidth=2, marker='o')

# Labeling the axes
plt.xlabel('Client number', fontsize=12)
plt.ylabel('Client condition', fontsize=12)

# Display the plot
plt.grid(True)
plt.title('Client Condition vs Client Number', fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuration for bandwidth
bw_low = 20
bw_high = 2500
number_of_clients = 50

# Generate random data for bandwidth, evenly spaced client numbers
bandwidths = np.random.randint(bw_low, bw_high + 1, size=number_of_clients)
clients = np.arange(1, number_of_clients + 1)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the data points with orange circles similar to the provided image style
plt.scatter(clients, sorted(bandwidths), s=100, edgecolor='orange', facecolor='none', linewidth=2, marker='o')

# Adding vertical bars
plt.axvline(x=number_of_clients / 3, color='blue', linestyle='--', linewidth=2)
plt.axvline(x=2 * number_of_clients / 3, color='blue', linestyle='--', linewidth=2)

# Labeling the axes
plt.xlabel('Client number', fontsize=12)
plt.ylabel('Client condition (Bandwidth)', fontsize=12)

# Display the plot
plt.grid(True)
plt.title('Client Condition vs Client Number', fontsize=14)
plt.savefig('D:\MS_Thesis\Pre-defense-june\Results_for_update\evenly_distributed_With_schemes.svg')

plt.show()
```
